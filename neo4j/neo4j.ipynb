```python
uri=os.getenv('NEO4J_URI') # Throws error while connecting need to change the schema
# Scheme	Meaning
# neo4j+s://	Encrypted + full certificate verification
# neo4j+ssc://	Encrypted + no certificate verification (trusts self-signed certs)
# neo4j://	No encryption (allows manual config via trusted_certificates)
```

```python
uri
```

```python
auth = (os.getenv("NEO4J_USERNAME"), os.getenv("NEO4J_PASSWORD"))
```

```python
auth
```

```python
uri = os.getenv('NEO4J_URI').replace('neo4j+s://', 'neo4j+ssc://')
```

```python
uri
```

```python
with GraphDatabase.driver(uri=uri, auth=auth) as d:
    d.verify_connectivity()
    session=d.session()
    result = session.run("RETURN 'Hello Vinay' AS message")
    print(result.single()["message"])
```
